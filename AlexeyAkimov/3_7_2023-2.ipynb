```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2.0*np.pi, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.plot(x, y*y)

plt.savefig("myplot.png")
```

```python

plt.figure(num=1, figsize=(16,8), dpi=300)
plt.subplot(1,2, 1)
plt.plot(x, y, label="sin", color="red", marker="", linestyle="--", linewidth=12)
plt.legend(fontsize=36)
plt.xlabel("x", fontsize=36)
plt.ylabel("sin(x)", fontsize=36)
plt.xlim( 0, 4)
plt.ylim( 0, 1.0)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


plt.subplot(1,2, 2)
plt.plot(x, y*y, label="$\sin^2$", color="#00FFFF", marker="o", markersize=12)
plt.legend(fontsize=36)
plt.xlabel("x", fontsize=36)
plt.ylabel("$sin^2(x)$", fontsize=36)
plt.xlim( 0, 4)
plt.ylim( 0, 1.0)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```
