```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
T = np.loadtxt('T.txt')
# props = ['E_kin', 'E_pot', 'E_tot', 'X', 'P']
# ylabels = ['Energy, a.u.', 'Energy, a.u.', 'Energy, a.u.', 'X, a.u.', 'P, a.u.']
E_tot = np.loadtxt('E_tot.txt')
E_pot = np.loadtxt('E_pot.txt')
E_kin = np.loadtxt('E_kin.txt')
plt.subplot(1,3,1)
plt.ylabel('Energy, a.u.')
plt.xlabel('Time')
plt.plot(T, E_kin, label='E_kin')
plt.plot(T, E_pot, label='E_pot')
plt.plot(T, E_tot, label='E_tot')
plt.legend()

plt.subplot(1,3,2)
X = np.loadtxt('X.txt')
plt.plot(X, T)
plt.xlabel('Time, a.u.')
plt.ylabel('X, a.u.')

plt.subplot(1,3,3)
P = np.loadtxt('P.txt')
plt.plot(X, P)
plt.xlabel('X, a.u.')
plt.ylabel('P, a.u.')

plt.tight_layout()
```
