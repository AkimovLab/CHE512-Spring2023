```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function1(x):
    return x
def function2(x):
    return np.sin(x)

def compute_integral(f, xmin, xmax, convergence_limit):
    """
    This function computes the integral of the function f for the interval of 
    [xmin, xmax].
    Ags:
        f (PyObject): The function
        xmin (float): The minimum x value
        xmax (float): The maximum x value
        convergence_limit (float): The convergence limit of the integration
    Returns:
        integral_value (float): The value of the integral
    """
    integral_value2 = 0
    # The number of points on the interval
    n = 1
    # Initializing the convergence value
    convergence = 1e2
    while convergence > convergence_limit:
        h = (xmax-xmin)/n
        xgrid = np.linspace(xmin,xmax,n)
        y = f(xgrid)
        integral_value1 = np.sum(y)*h
        convergence = np.abs(integral_value2-integral_value1)
        n += 1
        integral_value2 = integral_value1
        print(F'Iteration {n}, integral value is {integral_value1} with the convergence {convergence}.')
    return integral_value1

integral_value = compute_integral(function1, 10, 20, 1e-4)
integral_value = compute_integral(function2, 10, 20, 1e-5)
```
