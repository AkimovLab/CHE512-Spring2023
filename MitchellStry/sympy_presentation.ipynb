```python
import sympy as sp
```

```python
x=5
print (x)
```

```python
x,y,z=sp.symbols('x y z')
x=6*x
f=x*y*z
print(f)
print(x)
```

```python
x=sp.symbols('x')
#f(x)=2x+3
f=sp.Eq(2*x+3, 0)
f1=2*x+3
result=sp.solve(f1)
print(f)
print(result)
```

```python
x,y=sp.symbols('x y')
eq1=sp.Eq(2*x+y, 3) #2x+y=3
eq2=sp.Eq(2*x-y,4) #2x-y=5
res=sp.solve((eq1,eq2),(x,y))
print(res)
```

```python
#differentiation
x=sp.symbols('x')
f=2*x**3+sp.exp(-2*x)
result=sp.diff(f,x,3)
result2=sp.integrate(f,x)
result3=sp.integrate(f,(x,0,4))
print(result)
print(result2)
print(result3)
```
