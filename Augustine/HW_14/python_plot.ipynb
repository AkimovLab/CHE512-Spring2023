```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from files
T = np.loadtxt('tvalues.txt')
X = np.loadtxt('xvalues.txt')
P = np.loadtxt('pvalues.txt')
E_tot = np.loadtxt('Etot.txt')
E_pot = np.loadtxt('Epot.txt')
E_kin = np.loadtxt('Ekin.txt')

# Create figure and axes for position vs. time plot
fig1, ax1 = plt.subplots()

# Plot position vs. time
ax1.plot(T, X)

# Set labels and title
ax1.set_xlabel('Time')
ax1.set_ylabel('Position')
ax1.set_title('Position vs. Time')

# Create figure and axes for position vs. momentum plot
fig2, ax2 = plt.subplots()

# Plot position vs. momentum
ax2.plot(P, X)

# Set labels and title
ax2.set_xlabel('Momentum')
ax2.set_ylabel('Position')
ax2.set_title('Position vs. Momentum')

# Create figure and axes for energy vs. time plot
fig3, ax3 = plt.subplots()

# Plot energy components
ax3.plot(T, E_tot, label='Total Energy')
ax3.plot(T, E_pot, label='Potential Energy')
ax3.plot(T, E_kin, label='Kinetic Energy')

# Set labels and title
ax3.set_xlabel('Time')
ax3.set_ylabel('Energy')
ax3.set_title('Energy vs. Time')

# Add legend
ax3.legend()

# Show plots
plt.show()
```
