```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def merge_sort(inp_arr):
    size = len(inp_arr)
    if size > 1:
        middle = size // 2
        left_arr = inp_arr[:middle]
        right_arr = inp_arr[middle:]
 
        merge_sort(left_arr)
        merge_sort(right_arr)
 
        p = 0
        q = 0
        r = 0
 
        left_size = len(left_arr)
        right_size = len(right_arr)
        while p < left_size and q < right_size:
            if left_arr[p] < right_arr[q]:
              inp_arr[r] = left_arr[p]
              p += 1
            else:
                inp_arr[r] = right_arr[q]
                q += 1
             
            r += 1
 
        
        while p < left_size:
            inp_arr[r] = left_arr[p]
            p += 1
            r += 1
 
        while q < right_size:
            inp_arr[r]=right_arr[q]
            q += 1
            r += 1
            
    return inp_arr
 
```

```python
np.random.seed(1234)
linp_arr = [2, 4, 3,1, 5,92,34,24,25,255,2342,53]
print("Input Array:\n")
print(linp_arr)
merge_sort(linp_arr)
print("Sorted Array:\n")
print(linp_arr)
```

```python
def f(n):
    
    return n * np.log(n)

# Define the values of x
n_values = np.array([200, 400, 800, 1600, 3200, 6400, 12800, 25600])

# Compute the corresponding values of f(x)
y_values = f(n_values)

length_of_array = [200, 400, 800, 1600, 3200, 6400, 12800, 25600]
run_time = [169, 366, 808, 1760, 3850, 8020, 16900, 36300]  #The run time might differ a little after every run
```

```python
# Create two subplots side by side
fig, (bubble_sort, funct) = plt.subplots(1, 2, figsize=(10, 5))

# Plot f(x) on the left subplot
bubble_sort.plot(length_of_array, run_time)
bubble_sort.set_xlabel('Length of array')
bubble_sort.set_ylabel('time scale/microsec)')
bubble_sort.set_title('length of array vs run time using merge sort')

# Plot g(x) on the right subplot
funct.plot(n_values, y_values)
funct.set_xlabel('Length of array')
funct.set_ylabel('f(n) = n log(n))')
funct.set_title('Plot of f(n) = n log(n)')

# Adjust the spacing between the subplots
plt.subplots_adjust(wspace=0.3)

# Show the plots
plt.show()
```

```python
# We are now defining and plotting the bubble sort algorithm
def bubble_sort(inp_array):
    """
    Args:
        unsorted_list: (list of floats or ints): the list of the numbers to be sorted
        
    Returns:
        sorted_list : the sorted list
    """
    n = len(inp_array)

    # literating through all array elements
    for i in range(n):
        for j in range(n-i-1):          
            if inp_array[j] > inp_array[j+1]:
                inp_array[j], inp_array[j+1] = inp_array[j+1], inp_array[j]

    return inp_array
```
