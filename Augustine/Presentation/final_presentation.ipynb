```python
#1s orbital
import numpy as np
import matplotlib.pyplot as plt

a = 1  # Bohr radius
r = np.linspace(0, 5*a, 1000)  # Range of r values

# Probability density function
psi_squared = ( (1 / np.pi ) * ( (1/a)**3) * np.exp(-r/a))**2
# Radial probability density function
radial_prob_density = 4 * np.pi * r**2 * psi_squared

# Plotting the two graphs side by side
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# First plot for the probability density function
axs[0].plot(r, psi_squared)
axs[0].set_xlabel('r')
axs[0].set_ylabel('|psi(r)|^2')
axs[0].set_title('Probability density function for a 1s orbital')

# Second plot for the radial probability density function
axs[1].plot(r, radial_prob_density)
axs[1].set_xlabel('r')
axs[1].set_ylabel('P(r)')
axs[1].set_title('Radial probability density function for a 1s orbital')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from mpl_toolkits.mplot3d import Axes3D

# Define the function with two variables x and y
def f(x, y):
    return x * np.exp(-2 * np.sqrt(x**2 + y**2))

# Define the function with sympy
x, y = sp.symbols('x y')
f_sym = x * sp.exp(-2 * sp.sqrt(x * x + y * y))


# Generate a grid of x and y values
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)

# Calculate the corresponding z values
Z = f(X, Y)

# Create the first plot with contour lines
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 4))

ax1.contourf(X, Y, Z, levels=20, cmap='magma')
ax1.contour(X, Y, Z, levels=10, colors='black')
ax1.axhline(y=0, color='k')
ax1.axvline(x=0, color='k')
ax1.set_title('Contour plot of f(x,y) = x * exp(-2 * sqrt(x^2 + y^2))')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Evaluate the function with sympy and plot the surface
Z_sym = sp.lambdify([x, y], f_sym)(X, Y)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z_sym, cmap='coolwarm', edgecolor='none')
ax2.set_title('Isosurface Plot of f(x,y) = x * exp(-2 * sqrt(x * x + y*y))')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x,y)')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions with two variables x and y
def s_orbital(x, y):
    return np.exp(-2 * np.sqrt(x**2 + y**2))

def p_orbital(x, y):
    return x * np.exp(-2 * np.sqrt(x**2 + y**2))

def d_orbital(x, y):
    return (x*x - y*y) * np.exp(-2 * np.sqrt(x**2 + y**2))

def f_orbital(x, y):
    r = np.sqrt(x*x + y*y)
    f = (x*x - y*y) * (2*r / 3 - 1 ) * np.exp(-2 * r)
    return f

# Create a new figure with three subplots
fig, axs = plt.subplots(1, 4, figsize=(20, 4))

# Generate a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values for each function
Z0 = s_orbital(X, Y)
Z1 = p_orbital(X, Y)
Z2 = d_orbital(X, Y)
Z3 = f_orbital(X, Y)

# Plot each function with contour lines in a separate subplot
for i, ax in enumerate(axs.flat):
    if i == 0:
        ax.contourf(X, Y, Z0, levels=20, cmap='magma')
        ax.contour(X, Y, Z0, levels=10, colors='white')
        ax.set_title('S ORBITAL')
        ax.text(-4, 4, r'$f(x,y) = e^{-2 \sqrt{x^2 + y^2}}$', fontsize=15, color="white")
        
    elif i == 1:
        ax.contourf(X, Y, Z1, levels=20, cmap='magma')
        ax.contour(X, Y, Z1, levels=10, colors='black')
        ax.set_title('P ORBITAL')
        ax.text(-4, 4, r'$f(x,y) = x e^{-2 \sqrt{x^2 + y^2}}$', fontsize=15)
    elif i == 2:
        ax.contourf(X, Y, Z2, levels=20, cmap='magma')
        ax.contour(X, Y, Z2, levels=10, colors='black')
        ax.set_title('D ORBITAL')
        # Add the equation to the plot
        ax.text(-4, 4, r'$f(x,y) = (x^2 - y^2) e^{-2\sqrt{x^2 + y^2}}$', fontsize=15, color='black')
    elif i == 3:
        ax.contourf(X, Y, Z3, levels=20, cmap='magma')
        ax.contour(X, Y, Z3, levels=10, colors='black')
        ax.set_title('F ORBITAL')
        ax.text(3, 0, r'$r = \sqrt{x^2 + y^2}$',  fontsize=12)
        ax.text(3, -2, r'$f(x,y) = (x^2 - y^2)(2r/3-1)e^{-2r}$',  fontsize=15)
#     ax.axhline(y=0, color='k')
#     ax.axvline(x=0, color='k')
#     ax.set_xlabel('x')
#     ax.set_ylabel('y')

plt.show()
```
