```python
import numpy as np
#import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
def chaotic_motion_mapping(x0, r, n):
    """
    This function computes the recurrence relation of the form:
    xn+1 = 4rxn(1-xn)
    Args:
        x0 (float): The initial value of x
        r (float): The r value
        n (integer): The number of iteration to perform
    Returns:
        xn (float): The value of the function  nth iteration
    """
    iteration = 0
    
    while iteration < n:
   # if iteration < n:
        xn = 4*r*x0*(1-x0)
        x0 = xn
        iteration = iteration + 1
    return xn
```

```python
for jumps, runs in enumerate([10, 50, 100, 500]):
    plt.subplot(1,4,jumps+1)
    results = []
    x_range = np.arange(0.01,0.99,0.01)
    r_range = np.arange(0.1,0.99, 0.001)
    
    for x in x_range:
        for r in r_range:
            results.append([r, chaotic_motion_mapping(x,r,runs)])
    results = np.array(results)
    plt.scatter(results[:,0], results[:,1],  marker='.', color='black')
    plt.title(F'{runs} iterations')
    plt.xlabel('r_values')
    plt.ylabel('X iterations')
```
